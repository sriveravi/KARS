```python
# This script looks at the accuracy performance over the experiment 
# and makes nice learning curves

import pandas as pd
%pylab inline


# assumes you have run 'kars3b' and created the dataframe

df = pd.read_pickle( 'dfBehavKARSAll.p')
```

```python
#### SHOW INDIVIDUALS #########

# show supervised
subplot(121)
cond = 2
dfAdult = df[ (df.age > 30) &(df.condLab == cond)  ]
for s in dfAdult.subid.unique():
    thisS = dfAdult[dfAdult.subid ==s ]
    plot( thisS.trial, thisS.windAcc, linewidth=3)    
ylim([0,1])    
title('Adult supervised')

subplot(122)
cond = 0
dfAdult = df[ (df.age > 30) &(df.condLab == cond)  ]
for s in dfAdult.subid.unique():
    thisS = dfAdult[dfAdult.subid ==s ]
    plot( thisS.trial, thisS.windAcc, linewidth=3)
ylim([0,1])    
title('Adult Baseline')
```

```python
## Show adults on average

dfAdultAgg = df[ (df.age > 30)].groupby( [ 'condLab', 'trial']).mean().reset_index()
condList = ['BL', 'Lab', 'Sup']

for cond in [0,2]:    
    thisS = dfAdultAgg[dfAdultAgg.condLab ==cond ]
    plot( thisS.trial, thisS.acc, linewidth=3, label=condList[cond])    
ylim([0,1])    
title('Adult', fontsize=18)
legend(loc=4)


```
